## Compare subway traffic by day of the week. Display this as a graph.

```python
#Your code here
df.DATE = pd.to_datetime(df.DATE)
df['DayofWeek'] = df.DATE.dt.dayofweek
grouping = df.groupby('DayofWeek').sum()
grouping.plot(kind='bar')
```

## Is there more subway traffic on a weekend or a weekday?    Be specific in comparing magnitudes.

```python
#Your code here
grouping = grouping.reset_index()
grouping['Weekend'] = grouping.DayofWeek.map({0:False,1:False,2:False,3:False,4:False,5:True,6:True})
weekend_grouping = grouping.groupby('Weekend').mean()
weekend_grouping[['ENTRIES', 'EXITS']].plot(kind='bar')
#There appears to be equal subway traffic on weekdays and weekends : with entries at 1.2 x 10^12 and exits at 0.9 x 10^12.
```
